```python
# modules for different random numbers generation
#pseudo random numbers
import random

a = random.random()
print(a)
```

```python
import random

a = random.uniform(1,10)
print(a)
```

```python
import random

a = random.randint(1,10)
print(a)
```

```python
import random

a = random.randrange(1,10)
print(a)
```

```python
import random

a = random.normalvariate(0,1)
print(a)
```

```python
import random

mylist = list("ABCDEFGH")
print(mylist)

a = random.choice(mylist)
print(a)
```

```python
import random

mylist = list("ABCDEFGH")
print(mylist)

a = random.sample(mylist, 3)
print(a)
```

```python
import random

mylist = list("ABCDEFGH")
print(mylist)

a = random.choices(mylist, k=3)
print(a)
```

```python
import random

mylist = list("ABCDEFGH")
print(mylist)
random.shuffle(mylist)
print(mylist)
```

```python
#reproducing random numbers with random.seed
import random

random.seed(1)
print(random.random())
print(random.randint(1,10))

random.seed(2)
print(random.random())
print(random.randint(1,10))

random.seed(1)
print(random.random())
print(random.randint(1,10))

random.seed(2)
print(random.random())
print(random.randint(1,10))
```

```python
import secrets

a = secrets.randbelow(10)
print(a)

b = secrets.randbits(4)
print(b)

mylist = list("ABCDEFG")
c = secrets.choice(mylist)
print(c)
```
