```python
# different uses of * operator


result = 5 * 7
print(result)

```

```python
result = 2 ** 4
print(result)
```

```python
zeros = [0] * 10
print(zeros)
```

```python
zeros = [0, 1,2] * 10
print(zeros)
```

```python
zeros = (0, 1) * 10
print(zeros)
```

```python
zeros = "AB" * 10
print(zeros)
```

```python
def foo(a,b,*args,**kwargs):
  print(a,b)
  for arg in args:
    print(arg)
  for key in kwargs:
    print(key, kwargs[key])

foo(1, 2, 3,4,5,six=6,seven=7)
```

```python
def foo(a,b,*,c):
  print(a,b,c)

foo(1, 2,c=6)
```

```python
def foo(a,b,c):
  print(a,b,c)

my_list = [1,2,3]
foo(*my_list)
```

```python
def foo(a,b,c):
  print(a,b,c)

my_dict = {'a':1, 'b':2,'c':3}
foo(**my_dict)
```

```python
def foo(a,b,c):
  print(a,b,c)

my_dict = {'e':1, 'b':2,'c':3}
foo(**my_dict)
```

```python
numbers = [1,2,3,4,5,6]

*beginning, last = numbers
print(beginning)
print(last)
```

```python
numbers = (1,2,3,4,5,6)

*beginning, last = numbers
print(beginning)
print(last)
```

```python
numbers = (1,2,3,4,5,6)

beginning, *last = numbers
print(beginning)
print(last)
```

```python
numbers = (1,2,3,4,5,6)

beginning, *middle, last = numbers
print(beginning)
print(middle)
print(last)
```

```python
numbers = (1,2,3,4,5,6)

beginning, *middle, secondlast, last = numbers
print(beginning)
print(middle)
print(secondlast)
print(last)
```

```python
my_tuple = (1,2,3)
my_list = [4,5,6]

new_list = [*my_tuple, *my_list]
print(new_list)
```

```python
my_tuple = (1,2,3)
my_set = {4,5,6}

new_list = [*my_tuple, *my_set]
print(new_list)
```

```python
dict_a =  {'a':1,'b':2}
dict_b = {'c':3, 'd':4}
my_dict= {**dict_a, **dict_b}
print(my_dict)
```
