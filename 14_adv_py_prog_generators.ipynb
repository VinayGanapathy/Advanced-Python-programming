```python
# Generators

def countdown(num):
  print('Starting')
  while num > 0:
    yield num
    num -=1

cd = countdown(4)
print(cd)

value = next(cd)
print(value)

value = next(cd)
print(value)

value = next(cd)
print(value)

value = next(cd)
print(value)

value = next(cd)
print(value)
```

```python
# Generators are memory efficient
import sys

def firstn(n):
  nums = []
  num = 0
  while num < n:
    nums.append(num)
    num +=1
  return nums

mylist = firstn(20)
print(mylist)

print(sys.getsizeof(firstn(100000)))
```

```python
# Using generator
def firstn_generator(n):
  num = 0
  while num < n:
    yield num
    num +=1

print(list(firstn_generator(20)))

print(sys.getsizeof(list(firstn_generator(100000))))

print(sys.getsizeof(firstn_generator(100000)))

#print(sys.getsizeof(sum(firstn_generator(20))))
```

```python
# fibonacci as a generator

def fibonacci(limit):
  # 0 and 1 are initial numbers
  a , b = 0 , 1
  while a < limit:
    yield a
    a, b = b, a + b 

fib = fibonacci(30)

for i in fib:
  print(i)
```
