```python
# variable length arguments

def foo(a,b,*args,**kwargs):
  print(a, b)
  for arg in args:
    print(arg)
  for key in kwargs:
    print(key,kwargs[key])


foo(1,2)
```

```python
def foo(*args, last):
  for arg in args:
    print(arg)
  print(last)

foo(1,2,3)
```

```python
# unpacking arguments

def foo(a,b,c):
  print(a,b,c)

my_list=(0,1,2)
foo(*my_list)

my_list=[0,1,2]
foo(*my_list)
```

```python
# unpecking arguments

def foo(a,b,c):
  print(a,b,c)

my_list = [0,1,2,3,4]
foo(*my_list)
```

```python
# unpecking arguments

def foo(a,b,c):
  print(a,b,c)

my_dict = {'a':1,'b':2,'c':3}
foo(**my_dict)
```

```python
# unpecking arguments

def foo(a,b,c):
  print(a,b,c)

my_dict = {'a':1,'b':2,'c':3,'d':3}
foo(**my_dict)
```

```python
# Local vs global variables

def foo(): 
  x = number
  print('number inside function', x)


number = 0
foo()
```

```python
def foo(): 
  x = number
  number =3
  print('number inside function', x)


number = 0
foo()
```

```python
def foo(): 
  global number
  x = number
  number=3
  print('number inside function', x)


number = 0
foo()
print(number)
```

```python
def foo(): 
  number=3
  print('number inside function')


number = 0
foo()
print(number)
```

```python
def foo(): 
  global number
  number=3
  print('number inside function')


number = 0
foo()
print(number)
```

```python
# parameter parsing

def foo(x):
  x = 5

var = 10
foo(var)
print(var)
print(foo(var))
```

```python
def foo(a_list):
  a_list.append(4)

my_list = [1,2,3]
foo(my_list)
print(my_list)
print(foo(my_list))
```

```python
def foo(a_list):
  a_list.append(4)
  a_list[0]=-100

my_list = [1,2,3]
foo(my_list)
print(my_list)
print(foo(my_list))
```

```python
def foo(a_list):
  a_list = [200,300,400]
  a_list.append(4)
  a_list[0]=-100

my_list = [1,2,3]
foo(my_list)
print(my_list)
```

```python
def foo(a_list):
  a_list += [200,300,400]

my_list = [1,2,3]
foo(my_list)
print(my_list)
```

```python
def foo(a_list):
  a_list = a_list + [200,300,400]

my_list = [1,2,3]
foo(my_list)
print(my_list)
```
