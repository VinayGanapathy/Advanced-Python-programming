```python
# About Dictionaries : Key - value pairs, unordered, mutable
mydict = {"name" : "Max", "age": 28, "city" : "Newyork" }
print(mydict)
```

```python
mydict2 = dict(name = "Mary", age=27, city = "Boston")
print(mydict2)
```

```python
value = mydict["name"]
print(value)
```

```python
value = mydict["age"]
print(value)
```

```python
value = mydict["lastname"]
print(value)
```

```python
mydict["email"] = "max@xyz.com"
print(mydict)
```

```python
mydict["email"] = "coolmax@xyz.com"
print(mydict)
```

```python
del mydict["name"]
print(mydict)
```

```python
mydict.pop("age")
print(mydict)
```

```python
mydict.popitem()
print(mydict)
```

```python
mydict = {"name" : "Max", "age": 28, "city" : "Newyork" }
if "name" in mydict:
  print(mydict["name"])
```

```python
mydict = {"name" : "Max", "age": 28, "city" : "Newyork" }
try:
  print(mydict["name"])
except:
  print("Error")
```

```python
mydict = {"name" : "Max", "age": 28, "city" : "Newyork" }
try:
  print(mydict["lastname"])
except:
  print("Error")
```

```python
mydict = {"name" : "Max", "age": 28, "city" : "Newyork" }
for key in mydict:
  print(key)
```

```python
mydict = {"name" : "Max", "age": 28, "city" : "Newyork" }
for key in mydict.keys():
  print(key)
```

```python
mydict = {"name" : "Max", "age": 28, "city" : "Newyork" }
for value in mydict.values():
  print(value)
```

```python
mydict = {"name" : "Max", "age": 28, "city" : "Newyork" }
for key, value in mydict.items():
  print(key, value)
```

```python
mydict = {"name" : "Max", "age": 28, "city" : "Newyork" }
mydict_cpy = mydict

mydict_cpy["email"] = "max@xyz.com"

print(mydict)
print(mydict_cpy)
```

```python
mydict = {"name" : "Max", "age": 28, "city" : "Newyork" }
mydict_cpy = mydict.copy()

mydict_cpy["email"] = "max@xyz.com"

print(mydict)
print(mydict_cpy)
```

```python
mydict = {"name" : "Max", "age": 28, "city" : "Newyork" }
mydict_cpy = dict(mydict)

mydict_cpy["email"] = "max@xyz.com"

print(mydict)
print(mydict_cpy)
```

```python
mydict = {"name" : "Max", "age": 28, "city" : "Newyork" }
mydict2 = {"name" : "Max", "age": 28, "email" : "max@xyz.com" }

mydict.update(mydict2)
print(mydict)

```

```python
mydict = {3:9, 6:36, 9:81}
print(mydict)

value = mydict[3]
print(value)
```
