```python
my_list = ['a'] * 6
print(my_list)

my_string = ''
for i in my_list:
  my_string +=i
print(my_string)


#better method
my_string = ''.join(my_list)
print(my_string)
```

```python
from timeit import default_timer as timer
my_list = ['a'] * 6
print(my_list)

start = timer()
my_string = ''
for i in my_list:
  my_string +=i
stop = timer()
print(stop-start)


#better method
start = timer()
my_string = ''.join(my_list)
stop = timer()
print(stop-start)
```

```python
from timeit import default_timer as timer
my_list = ['a'] * 100000
print(my_list)

start = timer()
my_string = ''
for i in my_list:
  my_string +=i
stop = timer()
print(stop-start)


#better method
start = timer()
my_string = ''.join(my_list)
stop = timer()
print(stop-start)
```

```python
# formatting strings

# With % operator or .format() or f-strings

var = "Tom"
my_string = "the variable is %s" % var
print(my_string)

```

```python
var = 3
my_string = "the variable is %d" % var
print(my_string)
```

```python
var = 3.123456
my_string = "the variable is %f" % var
print(my_string)
```

```python
var = 3.123456
my_string = "the variable is %.2f" % var
print(my_string)
```

```python
var = 3.123456
my_string = "the variable is {}".format(var)
print(my_string)
```

```python
var = 3.123456
my_string = "the variable is {:.2f}".format(var)
print(my_string)
```

```python
var1 = 3.123456
var2 = 6
my_string = "the variable is {} and {}".format(var1, var2)
print(my_string)
```

```python
var1 = 3.123456
var2 = 6
my_string = f"the variable is {var1} and {var2}"
print(my_string)
```

```python
var1 = 3.123456
var2 = 6
my_string = f"the variable is {var1*2} and {var2}"
print(my_string)
```
