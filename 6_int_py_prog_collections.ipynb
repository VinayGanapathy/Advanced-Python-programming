```python
from collections import namedtuple

Point = namedtuple('Point','x,y')
pt = Point(1,-4)
print(pt)
print(pt.x,pt.y)
```

```python
from collections import OrderedDict
ordered_dict = OrderedDict()
ordered_dict['a'] = 1
ordered_dict['b'] = 2
ordered_dict['c'] = 4
ordered_dict['e'] = 3
ordered_dict['d'] = 5
print(ordered_dict)
```

```python
from collections import deque

d = deque()
d.append(1)
d.append(2)
d.append(2)
d.appendleft(3)
d.pop()
print(d)

d.popleft()
print(d)

d.extend([4,5,6])
print(d)

d.extendleft([4,5,6])
print(d)

d.rotate(2)
print(d)

d.rotate(-1)
print(d)
```
